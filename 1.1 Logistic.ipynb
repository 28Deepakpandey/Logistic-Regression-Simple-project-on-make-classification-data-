## Logistic Regression Implementation


````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
````

````python
# load dataset

from sklearn.datasets import make_classification
````

````python
x,y = make_classification(n_samples=1000,n_features=10, n_classes=2, random_state=15)
x,y
````

````python
## no need to do standardization

## Train_test_split


from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30, random_state=42)

````

````python
from sklearn.linear_model import LogisticRegression

logistic  = LogisticRegression()
logistic.fit(x_train,y_train)
y_pred = logistic.predict(x_test)
y_pred
````

````python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_pred,y_test)
````

````python
print(confusion_matrix(y_pred,y_test))
````

````python
print(classification_report(y_pred,y_test))
````

1. Direct Logistic Regression (without GridSearchCV):
Aap manually ek specific set of hyperparameters choose karte ho.

Model train hota hai ek hi parameter set ke saath.

Kya yeh parameter best hai? — Yeh aap sure nahi hote, unless aapne test karke dekha ho.

Overfitting ya underfitting ka risk hota hai.

 2. Logistic Regression with GridSearchCV:
 Model multiple combinations of hyperparameters ke saath train hota hai.

Har combination ke liye cross-validation apply hota hai.

Aapko best performing parameters milte hain automatically.

Result zyadatar cases me more accurate & generalized hota hai.

## Hyperparameter Tunning and Cross Validation 


## Grid SearchCV :ye ek hyperparameter tuning tool hai jo kisi machine learning model (jaise Logistic Regression, Random Forest, SVM, etc.) ke liye best parameters dhoondhne ke kaam aata hai.

| Feature | Direct Logistic Regression | With GridSearchCV |
|--------|----------------------------|--------------------|
| Parameters | Manually set | Automatically optimized |
| Accuracy | Depends on luck/guess | Usually better |
| Time | Fast | Slower (due to multiple model training) |
| Reliability | Low to Medium | High |

# RandomSearchCV

GridSearchCV aur RandomizedSearchCV dono hyperparameter tuning ke liye use hote hain, magar inmein kuch significant differences aur kuch common points hote hain:

### **Common Points:**
1. **Hyperparameter Tuning:** Dono ka use best hyperparameter combination dhundhne ke liye hota hai.
2. **Cross-Validation:** Dono methods cross-validation ka use karte hain taaki model performance accurately assess ki ja sake.
3. **Estimator:** Dono me kisi bhi sklearn estimator ka use kar sakte hain jaise ki RandomForestClassifier, SVC, etc.
4. **Parallel Processing:** Agar `n_jobs` parameter set karte hain, toh dono methods multiple cores ka use kar sakte hain computation speed badhane ke liye.

### **Differences:**
| Feature               | **GridSearchCV** | **RandomizedSearchCV** |
|-----------------------|-----------------|------------------------|
| **Search Method**     | Har possible hyperparameter combination ko systematically check karta hai. | Random sampling ke basis pe kuch limited combinations try karta hai. |
| **Computational Cost** | Zyada time-consuming hota hai, kyunki exhaustive search karta hai. | Kam computation time leta hai, kyunki sirf limited combinations check karta hai. |
| **Efficiency**       | Small datasets ke liye best hai, magar bade datasets ke liye slow ho sakta hai. | Large datasets ke liye better hai, kyunki randomly kuch specific hyperparameters pick karta hai. |
| **Best Use Case**     | Jab hyperparameters ke options limited aur predefined hote hain. | Jab hyperparameter space bada ho aur quick optimization chahiye. |

Agar dataset bada hai aur hyperparameter space wide hai, toh **RandomizedSearchCV** best rahta hai. Agar exact best configuration chahiye aur computing power ki koi limitation nahi hai, toh **GridSearchCV** suitable ho sakta hai.

---

## ✅ **1. Accuracy Score: `0.992`**

* Matlab: **99.2%** predictions correct hain.
* Lekin: Accuracy misleading ho sakti hai jab classes **imbalanced** ho (jaise yahaan pe 0 zyada hai, 1 bahut kam).

---

## ✅ **2. Confusion Matrix:**

```
[[2468    2]
 [  18   12]]
```

Yeh matrix ka matlab hai:

|               | **Predicted: 0** | **Predicted: 1** |
| ------------- | ---------------- | ---------------- |
| **Actual: 0** | 2468             | 2                |
| **Actual: 1** | 18               | 12               |

### Breakdown:

* **True Negatives (TN)** = 2468 → 0 ko sahi 0 predict kiya
* **False Positives (FP)** = 2 → 0 ko galti se 1 predict kiya
* **False Negatives (FN)** = 18 → 1 ko galti se 0 predict kiya
* **True Positives (TP)** = 12 → 1 ko sahi 1 predict kiya

---

## ✅ **3. Classification Report:**

```
              precision    recall  f1-score   support
```

Yeh metrics har class (0 aur 1) ke liye alag-alag bataye gaye hain.

### 🔹 Class 0:

* **Precision = 0.99** → Jitni baar 0 predict hua, 99% sahi the.
* **Recall = 1.00** → Jo actual 0 the, unme se 100% ko sahi pakda.
* **F1-score = 1.00** → Balanced score between precision & recall.
* **Support = 2470** → Test set me 0 class ke total samples.

### 🔹 Class 1 (minority class):

* **Precision = 0.86** → Jitni baar 1 predict hua, unme se 86% sahi the.
* **Recall = 0.40** → Jo actual 1 the, unme se sirf 40% sahi pakde.
* **F1-score = 0.55** → Not great, kyunki recall low hai.
* **Support = 30** → Test set me sirf 30 samples class 1 ke.

---

## ✅ **4. Macro & Weighted Average:**

### 🔸 Macro Avg:

* **Simple average** of both classes (0 and 1).
* Ignores imbalance.

```
Precision: (0.99 + 0.86)/2 = ~0.92  
Recall: (1.00 + 0.40)/2 = ~0.70  
F1-score: (1.00 + 0.55)/2 = ~0.77
```

### 🔸 Weighted Avg:

* Average weighted by support (number of samples).
* Yahaan pe mostly 0 class ka impact zyada hai.
* Isiliye accuracy aur weighted f1 high dikha raha hai.

---

## ⚠️ Kya Problem Hai?

* Aapka model **majority class (0)** ke liye bahut acha kaam kar raha hai.
* Lekin **minority class (1)** ke liye recall bahut low hai → sirf 40%.
* Iska matlab: model **bahut se important '1' cases miss kar raha hai**.

---

## ✅ Final Summary (Hinglish):

* Model **overall 99.2% accurate** hai, lekin class imbalance ki wajah se **class 1 ke liye weak** hai.
* Class 1 (e.g., fraud detection, disease, etc.) agar critical hai, to model ko **improve** karna zaroori hai.
* Aap **SMOTE, class\_weight='balanced'**, ya **threshold tuning** try kar sakte ho.

## ROC & AUC


 ROC aur AUC classification models ko evaluate karne ke **important tools** hain. Chaliye step-by-step simple hinglish mein samjhte hain:

---

## 🔹 **ROC Curve** – *Receiver Operating Characteristic*

### 🟢 Kya hai?

* **ROC ek graph hota hai**, jo batata hai ki model ka performance **different thresholds** par kaisa hai.
* X-axis: **False Positive Rate (FPR)**
* Y-axis: **True Positive Rate (TPR)** (*ya Sensitivity*)

### 🟡 Kyu hai?

* Jab aapke paas probability output ho (`predict_proba()`), to aap decide karte ho ki kis threshold par "1" bolna hai.
* ROC curve batata hai ki **har threshold par model ki sensitivity aur specificity** kaisi hai.

### 🛠️ Kaise banta hai?

1. Model se `predict_proba()` lo
2. Alag-alag threshold values (0 se 1 tak) par predictions banao
3. Har threshold par FPR aur TPR nikaalo
4. In values ka graph plot karo → milta hai **ROC Curve**

---

## 🔷 **AUC** – *Area Under the Curve*

### 🟢 Kya hai?

* **AUC** matlab ROC curve ke neeche ka **area**.
* Yeh ek **single value** deta hai (0 se 1 ke beech) jo overall model performance batata hai.

### 🟡 Kyu hai?

* Aapko har threshold ke performance ko visualize nahi karna padta — AUC ek hi number mein bata deta hai ki model **kitna achha** hai.

### 🔢 Interpretation:

| AUC Value | Meaning         |
| --------- | --------------- |
| 0.90–1.0  | Excellent model |
| 0.80–0.90 | Good            |
| 0.70–0.80 | Fair            |
| 0.50–0.70 | Poor            |
| 0.50      | Random guess    |

---

## 🔍 Example:

Agar aapka model `predict_proba()` deta hai:

```python
[0.9, 0.4, 0.7, 0.2, 0.6]
```

To aap different threshold (e.g., 0.5, 0.6, etc.) pe dekhoge ki:

* Kitne **true positive** aur **false positive** bane
* Fir inka ROC plot banta hai
* Us ROC ka area = **AUC**

---

### ✅ Summary (in short):

* **ROC**: graph of TPR vs FPR across thresholds
* **AUC**: us graph ke neeche ka area → ek number jo model performance batata hai
* **Higher AUC = better model**

Bilkul! Chaliye **"dummy model 0 value se kyu banate hain"** — is concept ko **step-by-step, bilkul simple aur hinglish** mein dobara samjhte hain. Aapne ROC-AUC ke context mein poocha hai, to usi angle se explain karte hain:

---

## 🔸 Step 1: Dummy Model kya hota hai?

Dummy model ek **naive (bewakoof) model** hota hai — jo bina soche-samjhe prediction karta hai.

### Example:

* Ek dummy model **hamesha 0 predict karta hai**

  * Chahe input kuch bhi ho — output sirf 0
* Ya kabhi-kabhi **random predict** karta hai (kabhi 0, kabhi 1)

---

## 🔸 Step 2: Dummy model ROC-AUC me kyu banate hain?

ROC-AUC curve banate waqt hum ek baseline (starting point) chahte hain:

> **"Kya hamara real model random se better hai?"**

Isliye hum ek dummy model banate hain jiska AUC **exactly 0.5** hota hai (ya uske aas-paas), taaki usse comparison ho sake.

---

## 🔸 Step 3: Ab "0 value se" banate hain ka kya matlab hai?

Yahaan iska matlab:

* Aapne ek dummy model banaya jo **hamesha class 0 predict karta hai**
* Iska matlab:

  * Sab samples ke liye `predict = 0`
  * Probability bhi sab ke liye `predict_proba = [1.0, 0.0]` (100% sure it's 0)

### Jab aap `roc_auc_score()` lagate ho:

* Positive class (1) ke liye **kabhi bhi 1 predict nahi hota**
* Iska ROC curve **bilkul neeche se seedha rehta hai**
* AUC \~ **0.5** hota hai (ya kabhi kabhi 0, depending on sklearn handling)

---

## 🔸 Step 4: Iska fayda kya?

Dummy model ka ROC-AUC:

* Aapko dikhata hai **minimum performance**
* Agar aapka real model bhi AUC ≈ 0.5 de raha hai, to iska matlab:

  > Model kuch nahi seekh raha — bas **random** ya **constant prediction** kar raha hai.

---

## 🔹 Ek chhoti analogy:

Sochiye aap ek exam de rahe ho jisme **True/False** questions hain.

* Dummy model = aap har question me bas “False” likh rahe ho (random nahi, sirf 0)
* Aapka score = 50% aa gaya (kabhi kabhi luck se)

Ab agar aapka real model bhi 50% score la raha hai, to kya yeh "intelligent" model hai? Nahi — bas dummy ke barabar hai.

---

## ✅ Final Summary:

| 🔍 Concept           | ✔️ Explanation                                                 |
| -------------------- | -------------------------------------------------------------- |
| Dummy Model          | Ek simple model jo hamesha 0 predict karta hai                 |
| 0 value se banana    | Dummy model sirf class `0` predict karta hai                   |
| ROC-AUC me kaam      | Compare karne ke liye — real model dummy se better hai ya nahi |
| AUC \~ 0.5 ka matlab | Model bas random ya fixed guess kar raha hai, kuch seekha nahi |

Aapka code bilkul sahi direction mein hai! Aap `roc_curve` function use kar rahe ho taaki **ROC curve** ke liye **FPR, TPR, aur thresholds** nikaal sako.

Aapka question hai — `thresholds` kya hai aur iska kya matlab hai?

---

## 🔸 `roc_curve` function kya return karta hai?

```python
fpr, tpr, thresholds = roc_curve(y_test, model_prob)
```

Ye function return karta hai:

| Output       | Meaning                                           |
| ------------ | ------------------------------------------------- |
| `fpr`        | False Positive Rate list (threshold ke hisaab se) |
| `tpr`        | True Positive Rate list (threshold ke hisaab se)  |
| `thresholds` | Har point par use kiya gaya threshold             |

---

## 🔸 Thresholds ka kya matlab hai?

* Jab model prediction karta hai, wo **probability** deta hai (e.g., 0.2, 0.8, 0.6...)
* Aap ek threshold fix karte ho — e.g., `0.5`
  Agar probability > 0.5 → predict 1
  Nahin to → predict 0

### Lekin ROC curve kya karta hai?

> Har possible threshold (0.0 se 1.0 tak) ke liye TPR aur FPR calculate karta hai
> Isiliye thresholds ka list milta hai, aur uske saath TPR/FPR change hote hain

---

### 🔹 Example:

| Threshold | TPR  | FPR  |
| --------- | ---- | ---- |
| 1.0       | 0.00 | 0.00 |
| 0.8       | 0.20 | 0.01 |
| 0.5       | 0.75 | 0.10 |
| 0.2       | 0.95 | 0.30 |
| 0.0       | 1.00 | 1.00 |

* Jab threshold low hota hai (0.0), to sabko 1 predict karta hai → TPR & FPR high
* Jab threshold high hota hai (1.0), to sabko 0 predict karta hai → TPR & FPR zero

---

## ✅ Summary:

* `thresholds` = probability cut-off points
* Unke basis par ROC curve ke liye TPR aur FPR bante hain
* Aap is list ko use karke decide kar sakte ho:

  * **Kis threshold par best trade-off mil raha hai?**
  * **Precision vs recall ka balance kahaan sahi hai?**
